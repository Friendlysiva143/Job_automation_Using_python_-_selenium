```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import time
from selenium.webdriver.chrome.service import Service
import pandas as pd
```

```python
driver=webdriver.Chrome(service=Service(ChromeDriverManager().install())) # To open the chrome browser
driver.get("https://www.naukri.com/") # To get the link in Chrome Browser
your_role=input("enter role") # job role you  want to apply
role=driver.find_element(By.CLASS_NAME,"suggestor-input ") # To get job title the search bar
role.send_keys(f"{your_role}") # send keys to the seach bar
role.send_keys(Keys.RETURN) # submit the keys in search bar
```

```python
data=[]
while True:
    # Print job details
    job_cards = WebDriverWait(driver,10).until(
        EC.presence_of_all_elements_located((By.CSS_SELECTOR, "div.srp-jobtuple-wrapper") )
    )#  selector for job card site
    for card in job_cards:
        # Job title + link
        try:
            title_elem = card.find_element(By.CSS_SELECTOR, "a.title")
            title = title_elem.text
            link = title_elem.get_attribute("href")
        except:
            title, link = "No title", "No link" 
        # Company name
        try:
            company = card.find_element(By.CSS_SELECTOR, "a[class*='comp-name']").text
        except:
            company = "No company"
        # Experience
        try:
            experience = card.find_element(By.CSS_SELECTOR, "span.exp-wrap").text
        except:
            experience = "No experience"
        
        # Salary
        try:
            salary = card.find_element(By.CSS_SELECTOR, "span.sal-wrap.ver-line").text
        except:
            salary = "No salary"
        try:
            # Get the <ul> container
            ul_elem = card.find_element(By.CSS_SELECTOR, "ul.tags-gt")
            # Now get all <li> inside that <ul>
            li_elems = ul_elem.find_elements(By.TAG_NAME, "li")
            # Extract text from each <li>
            skills = [li.text.strip() for li in li_elems if li.text.strip()]
            # Join skills into a single string
            skills_text = ", ".join(skills) if skills else "No skills"
        except:
            skills_text = "No skills mentioned"
        try:
            post_day=card.find_element(By.CSS_SELECTOR,"span.job-post-day").text
        except:
            post_day="1 day ago"
        if any(x in experience.lower() for x in ['fresher','no experiance']):
            min_exp=0
        else:
            try:
                min_exp = int(experience.split("-")[0].strip().split()[0])
            except:
                min_exp = -1
        if "1 day ago" in post_day:
            if min_exp==0:   
                # Print details
                print(f"Job Title: {title}")
                print(f"Apply Link: {link}")
                print(f"Company: {company}")
                print(f"Experience: {experience}")
                print(f"Salary: {salary}")
                print(f"posted Day:{post_day}")
                print(f"Skills:{skills_text}")
                print("-" * 50)
                data.append({"Job Title": title,"Apply Link": link,"Company": company,"Experience": experience,"Salary": salary,"Skills":skills_text})
    try:
        next_btn = WebDriverWait(driver, 5).until(
            EC.element_to_be_clickable((By.XPATH, "//a[span[text()='Next']]"))
        )
        driver.execute_script("arguments[0].click();", next_btn)
        time.sleep(3)  # wait for new page to load
    except:
        print("No more pages.")
        break
        driver.quit()
job_data=pd.DataFrame(data)
job_data.to_csv(f"{your_role}.csv",index=False)
print(f"Data saved in {your_role}.csv ")
```
